```python
import pandas
import numpy

examples = pandas.read_csv('adult_train.csv', header=None,
                       names=['age', 'capital-gain', 'capital-loss', 'hours-per-week', 'workClass', 'education', 'marital-status',
                              'occupation', 'relationship', 'race', 'sex', 'native-country', 'income'] )
print(examples.shape)  # Número de filas y columnas
examples.head(10)
```

```python
from sklearn import preprocessing
codificadores = []
examples_codificado = pandas.DataFrame()
for variable, valores in examples.iteritems():
    le = preprocessing.LabelEncoder()
    le.fit(valores)
    print('Codificación de valores para {}: {}'.format(variable, le.classes_))
    codificadores.append(le)
    examples_codificado[variable] = le.transform(valores)

examples_codificado.head(10)
```
